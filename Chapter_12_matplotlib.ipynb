###  선 그래프
- plt.plot([x,] y) x, y값에 대한 그래프 그리기
- plt.show() 그래프 출력
- plt.figure() 새로운 그래프 생성 
- plt.clf() 기존의 그래프 지우기
- plt.subplot(m,n,a) m x n로 그래프 나눠서 a위치에 출력
- plt.legend(['레전드이름']) 레전드 생성
- plt.xlim(a, b)  x축 a부터 b까지 범위 지정
- plt.ylim(a, b) y축 a부터 b까지 범위 지정 <br/><br/>
- %matplotlib qt 새로운 창에 그래프 출력
- %matplotlib inline 내부 출력창에 그래프 출력

**[12장: 292페이지] 그래프창 지정하기**   

```python
import numpy as np

# 데이터 생성
x = np.arange(-5, 5, 0.1)
y1 = x**2 -2
y2 = 20*np.cos(x)**2 # NumPy에서 cos()는 np.cos()으로 입력

plt.figure(1) # 1번 그래프 창을 생성함
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 2번 그래프 창을 생성함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(1) # 이미 생성된 1번 그래프 창을 지정함
plt.plot(x, y2) # 지정된 그래프 창에 그래프를 그림

plt.figure(2) # 이미 생성된 2번 그래프 창을 지정함
plt.clf() # 2번 그래프 창에 그려진 모든 그래프를 지움
plt.plot(x, y1) # 지정된 그래프 창에 그래프를 그림

plt.show()
```

**[12장: 294 ~ 295페이지] 하위그래프영역으로 나눈 후 각 영역에 그리기**

```python
import numpy as np

# 데이터 생성
x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)**2 + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2 # NumPy에서 sin()은 np.sin()으로 입력
y4 = 10*np.exp(-x) + 1 # NumPy에서 exp()는 np.exp()로 입력

# 2 × 2 행렬로 이뤄진 하위 그래프에서 p에 따라 위치를 지정
plt.subplot(2,2,1) # 2x2 그래프의 1번위치에 그림
plt.plot(x,y1)
plt.legend(['data1'])    # 레전드의 최적의 자리를 알아서 찾아줌

plt.subplot(2,2,2) # 2x2 그래프의 2번위치에 그림
plt.plot(x,y2)
plt.legend(['data1'])

plt.subplot(2,2,3) # 2x2 그래프의 3번위치에 그림
plt.plot(x,y3)
plt.legend(['data1'])

plt.subplot(2,2,4) # 2x2 그래프의 4번위치에 그림
plt.plot(x,y4)
plt.legend(['data1'])

plt.show()
```

**[12장: 311 ~ 312페이지]**

```python
import numpy as np

city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 위도(latitude)와 경도(longitude)
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]    # y축값 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]   # x축값

# 인구 밀도(명/km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]   # 마커의 크기 변수값

size = np.array(pop_den) * 0.8 # 마커 크기 변수 입력 (너무 커서 4/5크기로 줄임)
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 컬러 지정

plt.figure(figsize=(8,8))              # 출력 비율 설정
plt.scatter(lon, lat, s=size, c=colors, alpha=0.5)  # x, y값과 속성값을 각각 입력시킴, 알파값은 투명도
plt.xlabel('경도(longitude)')          # x축 이름
plt.ylabel('위도(latitude)')           # y축 이름
plt.title('지역별 인구 밀도(2017)')    # 타이틀 이름

#plt.xlim(126, 130)     # 출력범위 설정하기
#plt.ylim(34, 39)


                                       # 여러 변수를 반복변수로 함께 지정하기 위해 zip()함수 사용
for x, y, name in zip(lon, lat, city):
    plt.text(x, y, name)      # 그래프에 글씨를 씀, 위도 경도에 맞게 도시 이름 출력

plt.show()
```

### 막대 그래프

**[12장: 313페이지]**

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] # 회원 ID (그래프이름)
before_ex = [27, 35, 40, 33] # 운동 시작 전 횟수 (그래프1)
after_ex = [30, 38, 42, 37] # 운동 한 달 후 횟수 (그래프2)
```

**[12장: 313페이지]**

```python
import matplotlib.pyplot as plt
import numpy as np

n_data = len(member_IDs)     # 회원이 네 명이므로 전체 데이터 수는 4
index = np.arange(n_data)   # NumPy를 이용해 배열 생성 (0, 1, 2, 3)
plt.bar(index, before_ex)   # bar(x,y)에서 x=index, height=before_ex 로 지정
plt.show()
```

**[12장: 314페이지]**

```python
plt.bar(index, before_ex, tick_label = member_IDs)   # x축 숫자를 지우고 회원 이름 입력
plt.show()
```

**[12장: 315페이지]**

```python
colors=['r', 'g', 'b', 'm']   # 색 속성 리스트
plt.bar(index, before_ex, color = colors, tick_label = member_IDs)  # 색 속성값, 회원 이름 입력
plt.show()
```

**[12장: 315페이지]**

```python
plt.bar(index, before_ex, tick_label = member_IDs, width = 0.6)   # 너비 속성 변경
plt.show()
```

**[12장: 316페이지]**

```python
colors=['r', 'g', 'b', 'm']
plt.barh(index, before_ex, color = colors, tick_label = member_IDs)   # .barh 가로로 출력
plt.show()
```

**[12장: 317페이지]**

```python
barWidth = 0.4     # 바 너비 설정
plt.bar(index, before_ex, color='c', align='edge', width = barWidth, label='before')   # 그래프1
plt.bar(index + barWidth, after_ex , color='m', align='edge', width = barWidth, label='after')   # 그래프2

plt.xticks(index + barWidth, member_IDs)  # x라벨 태그 달기
plt.legend()
plt.xlabel('회원 ID')
plt.ylabel('윗몸일으키기 횟수')
plt.title('운동 시작 전과 후의 근지구력(복근) 변화 비교')
plt.show()
```
